# META DATA 통계 ( 두피 유형별 설문조사 결과 )

```python
import json
import os
import pandas as pd
from pandas import json_normalize
```

```python
def countAns(Fname):
    df = pd.DataFrame(columns=['q1','q2','q3','q4','q5','q6','q7'])
    file_list = os.listdir('./Training/'+Fname)
    id_list = []
    for file in file_list:
        with open(f'./Training/{Fname}/{file}','r') as f:
            json_d = json.load(f)
            id_list.append(json_d['image_id'])
    
    file_list = os.listdir('./Meta/META_DATA')
    for file in file_list:
        if file[:-13] in id_list:
            with open(f'./Meta/META_DATA/{file}','r') as f:
                json_d = json.load(f)
                new = pd.DataFrame([(json_d['answers1'],json_d['answers2'],json_d['answers3'],json_d['answers4'],json_d['answers5'],json_d['answers6'],json_d['answers7'])],columns=['q1','q2','q3','q4','q5','q6','q7'])
                df = df.append(new)
    print(Fname)
    for i in range(1,8):
        Q = 'q'+str(i)
        print('\n'+Q)
        print(df[Q].value_counts())
    
```

## 모낭사이홍반

### 양호

```python
countAns('[라벨]모낭사이홍반_0.양호')
```

### 경증

```python
countAns('[라벨]모낭사이홍반_1.경증')
```

### 중등도

```python
countAns('[라벨]모낭사이홍반_2.중등도')
```

### 중증

```python
countAns('[라벨]모낭사이홍반_3.중증')
```

## 모낭홍반농포

```python
for Lname in ['[라벨]모낭홍반농포_0.양호','[라벨]모낭홍반농포_1.경증','[라벨]모낭홍반농포_2.중등도','[라벨]모낭홍반농포_3.중증']:
    print('\n')
    countAns(Lname)
```
